```python
import pandas as pd
import tabula
import re
import psycopg2 as ps
```

```python
lista_tabelas = tabula.read_pdf("boletohotmart.pdf", pages="all", guess=False)
print(len(lista_tabelas))
```

```python
for tabela in lista_tabelas:
    
    tabela = lista_tabelas[0]
    tabela.columns = tabela.iloc[2]
    tabela =tabela.drop(0, axis=0)
    tabela = tabela.rename(columns={'Local de Pagamento Vencimento': 'index'})
    tabela = tabela.dropna(how="all")
    tabela.columns = tabela.loc[2,['index']]=['index']
    display(tabela)
    #print(tabela.loc[1, "Local de Pagamento Vencimento"])
```

```python
nr_linha_boleto = 1
nr_linha_identificacao = 20
nr_linha_valor = 9

boleto = tabela.loc[nr_linha_boleto, 'index']
print("\n",boleto)

boleto_num = re.findall(r'\b\d+\b', boleto)

aux_fd = []
aux_fd.append(boleto_num[0]) 
aux_fd.append(boleto_num[1]) 

bank_code = []
for i in aux_fd:
    bank_code.append(int(i))

outputBankCode = "".join(str(i) for i in bank_code)

print("\n Código do Banco:", outputBankCode)

del boleto_num[0]
del boleto_num[0]

boleto_num_df = []

outputBoletoNum = "".join(str(i) for i in boleto_num)
print("\n Número do Boleto: ", outputBoletoNum)

pagador = tabela.loc[nr_linha_identificacao,'index']
cpf =  re.findall(r'\b\d+\b', pagador)
outputCpf = "".join(str(i) for i in cpf)
print("\n Cpf:",outputCpf)

valor_aux = tabela.loc[nr_linha_valor,'index']
valor_df = [float(s) for s in re.findall(r'-?\d+\.?\d*', valor_aux)]
del valor_df[0]
del valor_df[1]

print("\n Valor: ",valor_df)
```

```python

def connect_to_db(host_name, dbname, port, username, password):
   try:
       conn = ps.connect(host=host_name, database=dbname, user=username,
       password=password, port=port)
 
   except ps.OperationalError as e:
       raise e
   else:
       print('Connected!')
       return conn
```

```python
host_name = 'localhost'
dbname = 'postgres'
port = '5432'
username = 'postgres'
password = '123456'
conn = None

conn = connect_to_db(host_name, dbname, port, username, password)
```

```python
curr = conn.cursor()
```

```python
def create_table(curr):
    create_table_command = ("""CREATE TABLE IF NOT EXISTS transactionsocr(
        cpf_cnpj VARCHAR PRIMARY KEY,
        cd_bank INTEGER NOT NULL,
        num_boleto VARCHAR(255) NOT NULL,
        valor NUMERIC NOT NULL
    )""")
    curr.execute(create_table_command)
```

```python
create_table(curr)
```

```python
conn.commit()
```

```python
def insert_into_table(curr, cpf_cnpj, cd_bank, num_boleto, valor):
    insert_into_transaction = ("""INSERT INTO transactionsocr (cpf_cnpj,  cd_bank,  num_boleto, valor)
    VALUES(%s,%s,%s,%s);""")

    row_to_insert = (outputCpf, outputBankCode, outputBoletoNum, valor_df)
    curr.execute(insert_into_transaction, row_to_insert)
```

```python
insert_into_table(curr, outputCpf, outputBankCode, outputBoletoNum, valor_df)
```
